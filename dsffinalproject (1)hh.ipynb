```python
# ============================================================
# PROJECT: Classification Modeling, Visualization & Comparison
# DATASET: Telco Customer Churn
# ============================================================

# =========================
# 1. IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# =========================
# 2. LOAD DATASET
# =========================
print("\n[TECHNIQUE: Dataset Loading]")
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
print("Dataset Loaded Successfully")
df.head()

# =========================
# 3. BASIC DATA EXPLORATION
# =========================
print("\n[TECHNIQUE: Initial Data Exploration]")
print("Dataset Shape:", df.shape)
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

# =========================
# 4. DATA PREPROCESSING
# =========================
print("\n[TECHNIQUE: Missing Value Handling & Type Conversion]")
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

print("\n[TECHNIQUE: Label Encoding Target Variable]")
le = LabelEncoder()
df['Churn'] = le.fit_transform(df['Churn'])

print("\n[TECHNIQUE: One-Hot Encoding Categorical Features]")
df_encoded = pd.get_dummies(df, drop_first=True)

# =========================
# 5. EXPLORATORY DATA ANALYSIS (EDA)
# =========================
print("\n[EDA: Churn Distribution – Understanding Class Balance]")
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()

print("\n[EDA: Histogram – Monthly Charges Distribution]")
df['MonthlyCharges'].hist(bins=30)
plt.title("Monthly Charges Distribution")
plt.show()

print("\n[EDA: Boxplot – Outlier Detection]")
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()

print("\n[EDA: Scatter Plot – Relationship Analysis]")
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.show()

print("\n[EDA: Correlation Heatmap – Numerical Features Only]")
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()

# =========================
# 6. FEATURE SCALING
# =========================
print("\n[TECHNIQUE: Feature Scaling using StandardScaler]")
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# 7. TRAIN / TEST SPLIT
# =========================
print("\n[TECHNIQUE: Train-Test Split]")
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
# =========================
# 8.  MODEL EVALUATION FUNCTION
# =========================
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# Initialize a list to store results for the final table
model_results = []

def evaluate_model(model, X_test, y_test, model_name, X_train=None, y_train=None):
    print(f"\n[MODEL OUTPUT: {model_name}]")

    # 1. Predictions
    # Handle ANN separately because it outputs probabilities
    if "ANN" in model_name:
        y_pred_prob = model.predict(X_test)
        y_pred = (y_pred_prob > 0.5).astype("int32")
    else:
        y_pred = model.predict(X_test)

    # 2. Calculate Basic Metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    # 3. Cross-Validation (Satisfies the Splitting/CV requirement)
    # We skip CV for ANN here to keep it simple, as Keras requires a wrapper
    cv_acc = "N/A"
    if X_train is not None and y_train is not None and "ANN" not in model_name:
        cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
        cv_acc = cv_scores.mean()
        print(f"Cross-Validation Accuracy (5-fold): {cv_acc:.4f}")

    # 4. Print Report & Plot Matrix
    print(classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    # 5. Store Results for Comparative Analysis
    model_results.append({
        "Model": model_name,
        "Accuracy": acc,
        "CV Mean Accuracy": cv_acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1
    })
# ==================================================
# 9. MODELS BEFORE pca
# ==================================================
print("\n========== MODELS BEFORE FEATURE ENGINEERING ==========")

# 1. Logistic Regression
print("\n[MODEL: Logistic Regression]")
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
evaluate_model(lr, X_test, y_test, "Logistic Regression (Base)", X_train, y_train)

# 2. Naïve Bayes
print("\n[MODEL: Naïve Bayes]")
nb = GaussianNB()
nb.fit(X_train, y_train)
evaluate_model(nb, X_test, y_test, "Naïve Bayes (Base)", X_train, y_train)

# 3. Decision Tree
print("\n[MODEL: Decision Tree]")
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
evaluate_model(dt, X_test, y_test, "Decision Tree (Base)", X_train, y_train)

# 4. Random Forest
print("\n[MODEL: Random Forest]")
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
evaluate_model(rf, X_test, y_test, "Random Forest (Base)", X_train, y_train)

# 5. ANN (Added validation_split to satisfy requirements)
print("\n[MODEL: Artificial Neural Network]")
ann = Sequential()
ann.add(Dense(16, activation='relu', input_shape=(X_train.shape[1],)))
ann.add(Dense(8, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Using validation_split=0.2 here satisfies the "Validation Set" requirement
ann.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)
evaluate_model(ann, X_test, y_test, "ANN (Base)")

# ==================================================
# 10. FEATURE ENGINEERING
# ==================================================
print("\n========== FEATURE ENGINEERING ==========")

print("\n[FEATURE ENGINEERING: Interaction Feature]")
df_encoded['Charges_Tenure'] = df_encoded['MonthlyCharges'] * df_encoded['tenure']

print("[FEATURE ENGINEERING: Binning Feature]")
df_encoded['TenureGroup'] = pd.cut(df['tenure'], bins=[0,12,24,48,72], labels=[1,2,3,4])

# =========================
# 10.1 HANDLE NaN BEFORE PCA
# =========================
print("[FEATURE ENGINEERING: Handling NaN Before PCA]")
imputer = SimpleImputer(strategy='median')
X_fe = imputer.fit_transform(df_encoded.drop('Churn', axis=1))

print("[FEATURE ENGINEERING: Principal Component Analysis (PCA)]")
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_fe)
# =========================
# 11. MODELS after pca
# =========================
print("\n========== MODELS AFTER FEATURE ENGINEERING (PCA) ==========")

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_pca, y, test_size=0.3, random_state=42
)

# 1. Logistic Regression (After PCA)
lr.fit(X_train_pca, y_train_pca)
# FIXED: Added X_train_pca and y_train_pca arguments
evaluate_model(lr, X_test_pca, y_test_pca, "Logistic Regression (After PCA)", X_train_pca, y_train_pca)

# 2. Random Forest (After PCA)
rf.fit(X_train_pca, y_train_pca)
# FIXED: Added X_train_pca and y_train_pca arguments
evaluate_model(rf, X_test_pca, y_test_pca, "Random Forest (After PCA)", X_train_pca, y_train_pca)

print("\nPROJECT EXECUTED SUCCESSFULLY")
print("\nPROJECT EXECUTED SUCCESSFULLY")
# =========================
# 12. FINAL COMPARATIVE ANALYSIS
# =========================
print("\n========== FINAL COMPARATIVE ANALYSIS ==========")

# Convert the results list to a DataFrame
results_df = pd.DataFrame(model_results)

# Sort by Accuracy to see the best model on top
results_df = results_df.sort_values(by="Accuracy", ascending=False)

# Display the table
print("\n[Summary Table of All Models]")
display(results_df)

# Visual Comparison
plt.figure(figsize=(12, 6))
sns.barplot(x="Accuracy", y="Model", data=results_df, palette="viridis")
plt.title("Model Accuracy Comparison")
plt.xlim(0, 1)
plt.show()

print("\nBest Performing Model based on Accuracy:")
print(results_df.iloc[0])
```
